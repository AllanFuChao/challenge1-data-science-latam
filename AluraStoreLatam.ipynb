## 📈Visualización de datos en grafica
Incluire gráficas claras y visuales</b> para facilitar la interpretación de los resultados actuales y ayudar a la toma de decisión para este proyecto.

<b>Se presenteran 3 graficos con información escencial para poder identificar cual de las tiendas es la mejor elección para cerrar.</b>

--Para cada grafica estaremos haciendo el desarrollo con la presentación de los datos y gráficos.--

### 1. Primeramente los ingresos totales de cada tienda:
Los ingresos totales nos ayudaran a identificar rapidamente cual tienda es rentable o no.

### 💵 Ingreso total por tienda

Según el análisis de facturación, **la Tienda 3** es la que genera **mayores ingresos totales**, mientras que **la Tienda 4** es la que presenta el ingreso más bajo.

```python
ingreso_total_tienda1 = round(tienda['Precio'].sum())
ingreso_total_tienda2 = round(tienda2['Precio'].sum())
ingreso_total_tienda3 = round(tienda3['Precio'].sum())
ingreso_total_tienda4 = round(tienda4['Precio'].sum())

ingresos = {
    'Tienda': ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4'],
    'Ingreso Total': [ingreso_total_tienda1, ingreso_total_tienda2, ingreso_total_tienda3, ingreso_total_tienda4]
}

df_ingresos = pd.DataFrame(ingresos)

plt.figure(figsize=(8, 5))
plt.bar(df_ingresos['Tienda'], df_ingresos['Ingreso Total'], color='seagreen')
plt.title('Ingreso Total por Tienda')
plt.xlabel('Tienda')
plt.ylabel('Ingreso Total ($)')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

### 2. Por segundo la calificación promedio de cada tienda
Tomando en cuenta que la popularidad de un negocio es importante.

### 🌟 Calificaciones promedio de clientes para cada tienda

Las calificaciones reflejan la experiencia del cliente. **La Tienda 3 tiene las mejores calificaciones promedio**, mostrando satisfacción general, mientras que **la Tienda 4** tiene una calificación menor, reflejando potenciales problemas en servicio, calidad o cumplimiento.

```python
promedio_tienda1 = round(tienda['Calificación'].mean(), 2)
promedio_tienda2 = round(tienda2['Calificación'].mean(), 2)
promedio_tienda3 = round(tienda3['Calificación'].mean(), 2)
promedio_tienda4 = round(tienda4['Calificación'].mean(), 2)

popularidad = {
    'Tienda': ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4'],
    'Promedio Ventas por Producto': [promedio_tienda1, promedio_tienda2, promedio_tienda3, promedio_tienda4]
}

df_popularidad = pd.DataFrame(popularidad)

plt.figure(figsize=(6, 6))
plt.pie(
    df_popularidad['Promedio Ventas por Producto'],
    labels=df_popularidad['Tienda'],
    autopct='%1.1f%%',
    startangle=90,
    colors=plt.cm.Pastel1.colors[:4]
)
plt.title('Popularidad Relativa por Tienda (Promedio de Ventas)')
plt.axis('equal')
plt.show()

```

### 3. Por ultimo el costo de envio en productos
Este dato nos ayuda a saber cual de las tiendas tienen mayor porcentaje de gasto en la distribución final de productos.

### ✉️ Costo de envío promedio

Respecto a la distribución, **la Tienda 2** presenta el **costo de envío más alto**, lo cual puede estar afectando su margen de ganancia. **La Tienda 3** mantiene un **costo de envío eficiente**, y **la Tienda 4** no compensa su bajo costo de envío con buen rendimiento general.

```python
prom_envio1 = round(tienda['Costo de envío'].mean(), 2)
prom_envio2 = round(tienda2['Costo de envío'].mean(), 2)
prom_envio3 = round(tienda3['Costo de envío'].mean(), 2)
prom_envio4 = round(tienda4['Costo de envío'].mean(), 2)

envios = {
    'Tienda': ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4'],
    'Costo Envío Promedio': [prom_envio1, prom_envio2, prom_envio3, prom_envio4]
}

df_envios = pd.DataFrame(envios)


plt.figure(figsize=(8, 5))
plt.barh(df_envios['Tienda'], df_envios['Costo Envío Promedio'], color='tomato')
plt.title('Costo de Distribución Promedio por Tienda')
plt.xlabel('Costo Promedio de Envío ($)')
plt.ylabel('Tienda')
plt.grid(axis='x', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()


```
